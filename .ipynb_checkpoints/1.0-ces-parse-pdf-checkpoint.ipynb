```python
import PyPDF2 as p2
import slate3k as slate
import re
```

```python
with open('p90x_nutrition_guide.pdf', 'rb') as f:
    pdf = slate.PDF(f)
```

```python
pages_with_recipe = []
for page in pdf:
    if 'serving:' in str(page):
        pages_with_recipe.append(page)
```

```python
titles = []
for page in pages_with_recipe:
    title = page.split('\n\n')[0]
    if len(title) > 1:
        titles.append(title)
```

```python
print(re.search(r"[0-9]", 'hello'))
```

```python
pattern = re.compile(r"L E V E L I I I\s{2}.+")
recipes = []
for page in pages_with_recipe:
    for recipe in re.split(pattern, page):
        recipes.append(recipe)
```

```python
for recipe in recipes:
    print(recipe.split("\n\n")[0:2])
    print("=========")
```
